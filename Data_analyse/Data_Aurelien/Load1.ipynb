```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns


base = Path.cwd().parents[1]
print(base)

```

# Nombre moyen d'accidents par tranches de 3 heures

```python
df = pd.read_csv(base/"Data_clean/full.csv")

df['heure'] = df['hrmn'].astype(str).str.zfill(4).str[:2].astype(int)

bins = list(range(0, 25, 3)) 
labels = [f"{i}h-{i+3}h" for i in range(0, 24, 3)]
df['tranche_3h'] = pd.cut(df['heure'], bins=bins, labels=labels, right=False, include_lowest=True)

accidents_par_tranche = df.groupby('tranche_3h').size()

plt.figure(figsize=(10,6))
accidents_par_tranche.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title("Nombre moyen d'accidents par tranche de 3 heures")
plt.xlabel("Tranche horaire")
plt.ylabel("Nombre d'accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv(base/"Data_clean/full.csv")

# Convertir hrmn en heure numérique
df['heure'] = df['hrmn'].str.split(':').str[0].astype(int)

# Définir les tranches horaires
bins = [0, 4, 8, 12, 16, 20, 22, 24]
labels = ['00-04', '04-08', '08-12', '12-16', '16-20', '20-22', '22-24']
df['tranche'] = pd.cut(df['heure'], bins=bins, labels=labels, right=False)

# Compter le nombre d'accidents par jour et tranche horaire
heatmap_data = df.groupby(['jour', 'tranche']).size().unstack(fill_value=0)

# Créer la heatmap
plt.figure(figsize=(10,6))
sns.heatmap(heatmap_data, annot=True, cmap="Reds", fmt="g")
plt.title("Nombre d'accidents par jour et tranche horaire")
plt.ylabel("Jour de la semaine")
plt.xlabel("Tranche horaire")
plt.show()
```
