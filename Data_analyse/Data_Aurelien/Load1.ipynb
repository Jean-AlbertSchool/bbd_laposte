```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt


base = Path.cwd().parents[1]
print(base)

```

# Nombre moyen d'accidents par tranches de 3 heures

```python
df = pd.read_csv(base/"Data_clean/full.csv")

df['heure'] = df['hrmn'].astype(str).str.zfill(4).str[:2].astype(int)

bins = list(range(0, 25, 3)) 
labels = [f"{i}h-{i+3}h" for i in range(0, 24, 3)]
df['tranche_3h'] = pd.cut(df['heure'], bins=bins, labels=labels, right=False, include_lowest=True)

accidents_par_tranche = df.groupby('tranche_3h').size()

plt.figure(figsize=(10,6))
accidents_par_tranche.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title("Nombre moyen d'accidents par tranche de 3 heures")
plt.xlabel("Tranche horaire")
plt.ylabel("Nombre d'accidents")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
